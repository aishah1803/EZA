```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send a GET request to the website
url = "https://www.retaildive.com/news/look-back-covid-19-impact-retail-pandemic/743853/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)

# Check the request was successful
if response.status_code == 200:
    print("Page successfully retrieved!")
else:
    print(f"Failed to retrieve page: {response.status_code}")

#  Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

#  Extracting the title
title_tag = soup.find('h1', {'class': 'display-heading-04'})
title = title_tag.get_text(strip=True) if title_tag else "No title found"

#  publication date
date_tag = soup.find('time')  # Finds the <time> tag
date = date_tag.get_text(strip=True) if date_tag else "No date found"

#  the author name
author_tag = soup.find('span', {'class': 'byline__author-name'})
author = author_tag.get_text(strip=True) if author_tag else "No author found"

#  article content
content_tag = soup.find('div', {'class': 'article-body'})
content = content_tag.get_text(separator="\n", strip=True) if content_tag else "No content found"

#  categories/tags
categories = [tag.get_text(strip=True) for tag in soup.find_all('a', {'class': 'article-category'})]
categories = ", ".join(categories) if categories else "No categories found"

#  image URLs
images = [img['src'] for img in soup.find_all('img') if 'src' in img.attrs]
image_urls = ", ".join(images) if images else "No images found"

#  all links from the article
links = [a['href'] for a in soup.find_all('a', href=True)]
links = ", ".join(links) if links else "No links found"

# Print results
print(f"Title: {title}")
print(f"Date: {date}")
print(f"Author: {author}")
print(f"Categories: {categories}")
print(f"Content Preview: {content[:600]}...")  # first 600 characters
print(f"Images: {image_urls}")
print(f"Links: {links}")

#  Save the data into a DataFrame
data = {
    "Title": [title],
    "Date": [date],
    "Author": [author],
    "Categories": [categories],
    "Content": [content],
    "Image URLs": [image_urls],
    "Links": [links]
}

#  CSV file
df = pd.DataFrame(data)
df.to_csv('retaildive_article2.csv', index=False)
print("\nData saved to 'retaildive_article2.csv'")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import re

# Fetch the webpage
url = "https://www.voguebusiness.com/story/technology/as-gen-z-matures-virtual-stores-get-an-ai-boost"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    print("Page successfully retrieved!")
else:
    print(f"Failed to retrieve page: {response.status_code}")

# Parse the page
soup = BeautifulSoup(response.text, 'html.parser')

# Extract metadata
title_tag = soup.find('title')
title = title_tag.get_text(strip=True) if title_tag else 'No title found'

date_tag = soup.find('meta', {'property': 'article:published_time'})
date = date_tag['content'] if date_tag else 'No date found'

author_tag = soup.find('meta', {'name': 'author'})
author = author_tag['content'] if author_tag else 'No author found'

# Extract content
content_tag = soup.find('article', {'class': 'article main-content'})
if content_tag:
    content = content_tag.get_text(separator="\n", strip=True)
else:
    content = 'No content found'

# Tokenize content into sentences
sentences = re.split(r'(?<=[.!?])\s+', content)

# Select random sentences
num_sentences = min(10, len(sentences))  # Get up to 10 random sentences
random_sentences = random.sample(sentences, num_sentences) if sentences else []

# Join selected sentences
content = " ".join(random_sentences)

# Extract images and links
image_urls = [img.get('src') for img in soup.find_all('img') if img.get('src')]
links = [link['href'] for link in soup.find_all('a', href=True)]

# Print extracted info
print(f"Title: {title}")
print(f"Date: {date}")
print(f"Author: {author}")
print(f"Content Sample: {content[:600]}...")
print(f"Images: {image_urls[:3]}...")
print(f"Links: {links[:3]}...")

# Save to CSV
data = {
    "Title": [title],
    "Date": [date],
    "Author": [author],
    "Content": [content],
    "Images": [', '.join(image_urls)],
    "Links": [', '.join(links)]
}

df = pd.DataFrame(data)
df.to_csv('vogue_business_article_random.csv', index=False)
print("\nData saved to 'vogue_business_article_random.csv'")
```
