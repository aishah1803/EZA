```python
from pydub import AudioSegment

def convert_mp3_to_wav(mp3_file, wav_file):
    # Convert the MP3 file to WAV format
    audio = AudioSegment.from_file(mp3_file, format="mp3")
    audio.export(wav_file, format="wav")
    print(f"Converted {mp3_file} to {wav_file}")

# Convert the first podcast file
convert_mp3_to_wav("WSMM6416199842.mp3", "podcast1.wav")
```

```python
import speech_recognition as sr
from pydub import AudioSegment

def transcribe_audio_in_chunks(wav_file, chunk_length=30):
    """
    Transcribes an audio file by sending 30-second chunks to Google's Speech-to-Text API.
    """
    recognizer = sr.Recognizer()

    # Load the full audio file
    audio = AudioSegment.from_wav(wav_file)

    # Calculate total duration in seconds
    total_duration = len(audio) / 1000
    print(f"Total audio duration: {total_duration:.2f} seconds")

    full_transcription = ""

    # Process the audio in 30-second chunks
    for start_time in range(0, int(total_duration), chunk_length):
        end_time = min(start_time + chunk_length, total_duration)

        print(f"Processing segment: {start_time}s to {end_time}s")

        # Extract the chunk
        chunk = audio[start_time * 1000:end_time * 1000]  # Convert seconds to milliseconds
        chunk_filename = f"chunk_{start_time}_{end_time}.wav"
        chunk.export(chunk_filename, format="wav")

        # Transcribe the chunk
        with sr.AudioFile(chunk_filename) as source:
            audio_data = recognizer.record(source)

        try:
            chunk_text = recognizer.recognize_google(audio_data)
            print(f"Chunk transcription ({start_time}s - {end_time}s): {chunk_text}\n")
            full_transcription += chunk_text + " "
        except Exception as e:
            print(f"Error transcribing chunk {start_time}s - {end_time}s: {e}")

    return full_transcription

# Run the function on your WAV file
full_text = transcribe_audio_in_chunks("podcast1.wav")
print("\nFinal Full Transcription:\n", full_text)
```

```python
# Save the full transcription into a text file
with open("full_transcription.txt", "w") as f:
    f.write(full_text)

print("Transcription saved to full_transcription.txt")

# Optionally, download the file to your local machine
from google.colab import files
files.download("full_transcription.txt")
```
